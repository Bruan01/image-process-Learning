## 点运算
点运算
- 只改变本身像素
- 不会改变其空间位置


## 导包
### 图像处理包

```python
from skimage import data,io,exposure
```

### 绘图包

```python
from matplotlib import pyplot as plt
```

```python
# 读入图像
image = data.coffee()
# 分别计算 gamma 不同时对模型产生的情况
# adjust_gamma 可以改变 图像处理格式中的gamma值 (点运算) 从而使图像发生灰度变化
image_1 = exposure.adjust_gamma(image,0.2)
image_2 = exposure.adjust_gamma(image,0.76)
image_3 = exposure.adjust_gamma(image,23)
```

```python
# show image
plt.subplot(2,2,1)
plt.title("gamma = 1")
io.imshow(image)
plt.subplot(2,2,2)
plt.title("gamma = 0.2")
io.imshow(image_1)
plt.subplot(2,2,3)
plt.title("gamma = 0.76")
io.imshow(image_2)
plt.subplot(2,2,4)
plt.title("gamma = 23")
io.imshow(image_3)
plt.show()
```

## 参数查找展示
### 图像直方图
- 统计颜色出现的频数
- 双峰特性： 出现在背景与前置不耦合的情况

```python
from skimage import exposure,data
image  = data.coffee()
# 计算直方图
hist_r = exposure.histogram(image[:,:,0],nbins=256)
hist_g = exposure.histogram(image[:,:,1],nbins=256)
hist_b = exposure.histogram(image[:,:,2],nbins=256)
```

```python
# 绘图
from matplotlib import pyplot as plt
import numpy as np
# show image
plt.subplot(2,2,1)
plt.title("原图")
io.imshow(image)
print(type(image),
type(hist_r))
print(hist_r)
hist_r = np.array(hist_r)
plt.subplot(2,2,2)
plt.title("R")
io.imshow(hist_r)
plt.subplot(2,2,3)
plt.title("G")
plt.show(hist_g)
plt.subplot(2,2,4)
plt.title("B")
plt.show(hist_b)
plt.show()
```
