## 图像的表示与描述
- 1.了解图像特征在空间域和频率域的表示和描述方法；  
- 2.掌握图像特征表示和描述的Python实现方法。

### 1.彩色和灰色图的直方图

```python
import cv2
import matplotlib.pyplot as plt 
img = cv2.imread("./thfind.jpg")
gray=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
img_rgb=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
fig,axes = plt.subplots(1,4,figsize=(15,8),dpi=200)
axes[0].imshow(img)
axes[0].set_title("origin")
axes[1].hist(img_rgb.ravel(),256)
axes[1].set_title("hist_color")
axes[2].imshow(gray,"gray")
axes[2].set_title("gray")
axes[3].hist(gray.ravel(),256)
axes[3].set_title("hist_gray")
plt.show()
```

### 形态学描述
1. 图像腐蚀和膨胀  
 cv.erode(img, kernel) 腐蚀  
 cv.dilate(img,kernel)  膨胀

2. 骨架化

```python
import numpy as np
import cv2 as cv 
import matplotlib.pyplot as plt
# 1. 读取图像
img = cv.imread("./boy.jpg")
# cv.cvtColor(img,cv2.COLOR_RGB2BGR)

# 2 创建核结构
kernel = np.ones((20,20), np.uint8)
# 3 图像腐蚀和膨胀
erosion = cv.erode(img[:,:,::-1], kernel) # 腐蚀
dilate = cv.dilate(img[:,:,::-1],kernel) # 膨胀

flg,axes = plt.subplots(1,3,dpi=200,figsize=(12,8))
axes[0].imshow(img[:,:,::-1])
axes[0].set_title("origin")
axes[1].imshow(erosion)
axes[1].set_title("erosion")
axes[2].imshow(dilate)
axes[2].set_title("dilate")

plt.show()


```

（2）图像边界提取与骨架化  
Canny边缘检测：cv2.Canny  
骨架化：imutils.skeletonize  

```python
import cv2
import imutils
import numpy as np
import matplotlib.pyplot as plt
img=cv2.imread('./black_boy.jpg')
gray=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
canny=cv2.Canny(gray,80,150)
# 边界提取图像
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(10,8),dpi=600)
axes[0].imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
axes[0].set_title("orign")
axes[1].imshow(canny,'gray')
axes[1].set_title("Canny detection")
plt.show()
```

骨架化图像：
骨架图依赖灰度图
骨架化图像
先进行膨胀后进行骨架化

```python
# 定义卷积核
kernel=np.ones((1,1),np.uint8)
dilate=cv2.dilate(canny,kernel)# 膨胀
skeleton = imutils.skeletonize(dilate, size=(3,3)) #骨架
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(10,8),dpi=800)
axes[0].imshow(dilate,'gray')
axes[0].set_title("canny")
axes[1].imshow(skeleton,'gray')
axes[1].set_title("skeleton")
plt.show()
```
