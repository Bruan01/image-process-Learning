## 主要操作命令
### （1）. cv2.imread    图像文件的读取

```python
import cv2
plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签
# 0 表示单通道
I = cv2.imread('cameraman.png',0)
# 文件名 + 图片对象
cv2.imshow("cameraman",I)
cv2.waitKey(0)
```

### （2）. cv2.imwrite   图像文件的写入
在工作文件下面会生成一个名为cameraman的png格式的图像文件

```python
import cv2
img = cv2.imread('cameraman.png')
cv2.imwrite('cameraman.png',img)
```

### （3）.os.stat 查询图像文件信息

```python
import os 
fileinfo = os.stat("cameraman.png")
print("索引号：",fileinfo.st_ino)
print("被连接数目：",fileinfo.st_dev)
print("文件大小:",fileinfo.st_size,"字节")
print("最后一次访问时间:",fileinfo.st_atime)
print("最后一次修改时间:",fileinfo.st_mtime)
print("最后一次状态变化的时间：",fileinfo.st_ctime)
```

### （4）. cv2.imshow  
显示图像

```python
import cv2
I = cv2.imread('cameraman.png')
cv2.imshow("cameraman",I)
cv2.waitKey(0)
# 设置 waitKey(0) , 则表示程序会无限制的等待用户的按键事件
# 一般在 imgshow 的时候 , 如果设置 waitKey(0) , 代表按任意键继续
```

### （5）.一个图像窗口中显示多幅图像


```python
import cv2
import matplotlib.pyplot as plt
img3 = cv2.imread('beyond.png')
gray=cv2.cvtColor(img3,cv2.COLOR_RGB2GRAY) #灰度化
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(10,8),dpi=100)
# img[:,:,::-1]的作用是列表数组左右翻转
# 我们用cv2的imshow()函数显示图片发现跟我们打开图片的样子一样，
# 是因为cv2的imshow()又把BGR转回RGB再显示。
# 但cv2确实是以BGR形式读取图片的，而plt则是以RGB形式。
axes[0].imshow(img3[:,:,::-1])
axes[0].set_title("orign pic")
axes[1].imshow(gray[:-1])
axes[1].set_title("grey")
plt.show()
```

### （6）.cv2.add     两幅图像相加

```python
import cv2 
img1 = cv2.imread('cameraman.png')

img3 = cv2.add(img1,img1) # cv加法
cv2.imshow('origin',img1)
cv2.imshow('imgname',img3)
cv2.waitKey(0)
```

### （7）. subtract  将一幅图像从另一幅图像中减去

```python
import cv2
img1 = cv2.imread('cameraman.png');
img3 = cv2.subtract(img1,img1) # cv加法
cv2.imshow('origin',img1)
cv2.imshow('subtract',img3)
cv2.waitKey(0)
```

### （8）.resize     图像的缩放

```python
# 代码：
img1=cv2.imread('cameraman.tif');
# 2.1 绝对尺寸
# rows,cols = img1.shape[:2]
print(img1.shape)
rows,cols = img1.shape[:2]
cv2.imshow('img1',img1)
res = cv2.resize(img1,(2*cols,2*rows),interpolation=cv2.INTER_CUBIC)
cv2.imshow('res',res)
# cv2.waitKey(0)
# 2.2 相对尺寸
res1 = cv2.resize(img1,None,fx=0.5,fy=0.5)
cv2.imshow('res1',res1)
cv2.waitKey(0)
```

### （9）.warpAffine图像的旋转

```python
import numpy as np
import cv2 
import matplotlib.pyplot as plt
# 读取图像
img = cv2.imread('cameraman.png')
# 图像旋转
rows,cols = img.shape[:2]
# 生成旋转矩阵  旋转中心点坐标  角度 缩放比例
M = cv2.getRotationMatrix2D((cols/2,rows/2),90,1)
# 旋转变换
dst = cv2.warpAffine(img,M,(cols,rows))
# 3 图像展示
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(5,8),dpi=100)
axes[0].imshow(img[:,:,::-1])
axes[0].set_title("原图")
axes[1].imshow(dst[:,:,::-1])
axes[1].set_title("旋转后结果")
```

### （10）. 图像的剪切

```python
img = cv2.imread('cameraman.png')
cat = img[50:160,45:115]
cv2.imshow("imgname",cat)
cv2.waitKey(0)
```

3.读入一幅图像（自选），分别将其顺时针旋转30度，逆时针旋转30度，并将原图和旋转后的图显示在一个图形框（总共3幅图）。

```python
    import cv2 
    import matplotlib.pyplot as plt
    img = cv2.imread('beyond.png')
    rows,cols = img.shape[:2]
    # 生成逆旋转矩阵
    img_ccw = cv2.getRotationMatrix2D((cols/2,rows/2),30,1)
    # 生成顺旋转矩阵
    img_cw = cv2.getRotationMatrix2D((cols/2,rows/2),-30,1)
    # 旋转变换
    ccw = cv2.warpAffine(img,img_ccw,(cols,rows))
    cw = cv2.warpAffine(img,img_cw,(cols,rows))
    # 3 图像展示
    fig,axes=plt.subplots(nrows=1,ncols=3,figsize=(10,8),dpi=100)
    axes[0].imshow(img[:,:,::-1])
    axes[0].set_title("原图")
    axes[1].imshow(ccw[:,:,::-1])
    axes[1].set_title("逆时针30旋转后结果")
    axes[2].imshow(cw[:,:,::-1])
    axes[2].set_title("顺时针30旋转后结果")
```
